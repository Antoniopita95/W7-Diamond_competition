```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
import pandas as pd
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv("data/clean_train_01.csv")
test = pd.read_csv("data/clean_test_01.csv")
train.head()
```

```python
y = train["price"]
X = train.drop(columns=["price"])
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=10)
```

## Random Forest Regressor

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=None, max_features='auto')
#entrenamos el modelo
rf.fit(X_train, y_train)

# metrica
y_pred_train = rf.predict(X_train)
y_pred_test = rf.predict(X_test)

# Metrics
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```

```python
# define model and params

rf_model = RandomForestRegressor() 

params = {
    "n_estimators":[50, 100, 200], 
    "max_depth":[5, 20, None], 
    "min_samples_split":[2, 4], 
    "min_samples_leaf":[1, 2]
}
```

```python
mse = make_scorer(mean_squared_error)
```

```python
# run grid search

clf = GridSearchCV(estimator = rf_model, param_grid = params, scoring = mse, verbose=2)
```

```python
# fit model

clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
# define best model

best = clf.best_estimator_
best
```

```python
# predict using best model

y_pred_train = best.predict(X_train)
y_pred_test = best.predict(X_test)
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)
mse_train, mse_test
```
